```python
#Initiliazing Libraries for Data Analysis
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#This Dataset comes from kaggle and is a dataset of risk factors for Cervix Cancer.
data = pd.read_csv("C:/Users/lench/Desktop/Cleaning_Dataset/kag_risk_factors_cervical_cancer.csv")
data.head()
```

```python
#Deploying dataset relevant information
print(f'\nNumber of rows:{data.shape[0]}')
print(f'Number of columns:{data.shape[1]}')
print('Types of data in the dataset:')
print(data.dtypes)
```

```python
#Change Object to Numeric in the dataset variables.
objects = data.select_dtypes(include=['object']).columns
data[objects] = data[objects].apply(pd.to_numeric, errors='coerce', downcast='integer')
print('Data set after changing object to numeric:')
data.describe()
```

```python
print('Data Null Values per Variable:')
data.isnull().sum()
```

```python
#Replacing Data Null Values with the Median of the Dataset
data.fillna(data.median(), inplace=True)
```

```python
#Checking if the Null Values have been replaced
print('Data Null Values per Variable after replacing with median:')
data.isnull().sum()
```

```python
data.to_csv('CleanData.csv', index=False)
```
